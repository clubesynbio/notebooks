##Simulation

To obtain the behavior of our model described at our [Modeling page](http://2013.igem.org/Team:USP-Brazil/Modeling) we codefied the model final equation:

$$ 
[R]_{eq} = k_{et} \frac{1}{1+\lambda \frac{[met]}{et}} + k_{met} \frac{1}{1+\lambda^{-1} \frac{[et]}{met}}
$$

In order to see how it behaves we need to computate this equation for each possible combination of $[et], [met]$ in a certain range.Also, we need to estimate values for the parameters $k_{et}$, $k_{met}$ and $\lambda $, task that required the use of experimental data.

We begin by importing the relevant modules:

+ Numpy is a module to work with numeric array in python
+ Matplotlib is a module to plot graphs of functions. Here we require only the submodule pyplot

```python
import numpy as np
import matplotlib.pyplot as plt
# next line is a magic to displya graph 'inline' within the notebook
%matplotlib inline
```

Next we code the equation for $R_{eq}$ 

```python
def R_eq(lambda_,k_et,k_met):
    """
    R, R_max, R_min, met, et = R_eq(lambda_, k_et, k_met).
    
    Generate a grid os values for [et], [met] varing in the interval [0,1]
    and compute the values of [R]_eq at each point. Return R, R_max, R_min, met and et.
    
    :lambda_:
        positive float
    :k_et:
        positive float
    :k_met:
        positive float
        
    
    """
    
    delta_et, delta_met = 1.e-3, 1.e-3  # refinement for the grid
    
    et, met = np.mgrid[slice(1.e-4, 1 + delta_et, delta_et),     # grid with all possible
                       slice(1.e-4, 1 + delta_met, delta_met)]   # combinations of relative
                                                                 # ethanol and methanol.
    
    lambda_et = lambda_ * (met / et)          # auxiliar variable
    lambda_met = 1/(lambda_ * (et / met))     # auxiliar variable
    
    R = k_et * (1 / (1 + lambda_et)) + k_met * (1 / (1 + lambda_met))
    
    R = R[:-1,:-1]  # excluding borders to ensure R is inside the boundaries of met, et
    
    R_min, R_max = np.min(R), np.max(R)

    return R, R_max, R_min, met, et  
```

Next we need to set the parameter. To do so, we used the the single point of data available regarding the behavior of Pichia pastoris RFP expression with the $P_{AOX}$ and the modified transcription factor $MXR1-S215A$, found at [1](#Ppastoris).

<img src="https://dl-web.dropbox.com/get/Public/PpastorisDataGraph.png?w=AAB3J0xrBDxG35ZyvF_Jc_STEX95oQR4D9mE08NyMIfU4A">

This graph displays the expresion of $P_{AOX}$ in different relative concentrations of ethanol and methanol. The color white represents a medium with $0.5\%$ ethanol only, the color black is for $0.5\%$ methanol only and the gray is for both methanol and ethanol at $0.5\%$. The relevant data is for $MXR1-S215A$.

Also, looking at the formula for $R_{eq}$ we can see that, when ethanol or methanol is not present, the value for the remaining $k$coefficient can be determined by the value of the expression alone. So, we can easily estimate the parameters:

```python
k_et = 0.1
k_met = 4.101
lambda_ = (k_et - 1.837) / (1.837 - k_met)
print "lambda = %f" % lambda_


RFP, MAX, MIN, MET, ET = R_eq(lambda_,k_et,k_met)
print "Max RFP concentration = %f\nMin RFP concentration = %f" % (MAX, MIN)
```

```python
plt.pcolor(MET, ET, RFP, cmap='gist_heat', vmin=MIN, vmax=MAX)
plt.title('RFP concentrarion x Methanol concentration x Ethanol concentration')
plt.axis([0, 1, 0, 1])
plt.xlabel('[met]')
plt.ylabel('[et]')
plt.colorbar()
```

It is worth to note that, despite the data was obtained for the expression in unities of mRNA, the molar concentration of RFP is directly related to this value.

## Conclusion

We can see from the color graph that the RFP expression is drastically reduced, compared to the full capacities of P. pastoris without the inhibition/competition put by ethanol. Also, for our methanol detector to be useful in a alcoholic drink, the operacional conditions are always bellow the diagonal, and we can conclude through this analysis that we can expect no more than half the expression in operation conditions compared to ideal situation for P. pastoris. Actualy the condition for "almost" commercial drink stated by OMS says that we will be looking for at most $2\%$ of volume in methanol. Common drink have ethanol volumes varing from $4\%$ to $40\%$ so there are situations in the produtc's operationa condition where nearly zero RFP expression is expected.
